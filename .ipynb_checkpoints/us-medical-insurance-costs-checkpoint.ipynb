# U.S. Medical Insurance Costs

## Goal 

My goal from this project is to find out what is the percent of young aged people (up to 21) which have insurance cost higher than median of the whole dataset.

```python
import csv
from decimal import Decimal

insurance_dict = {}
with open('insurance.csv', 'r', newline = '') as insurance_csv:
    insurance_dict = csv.DictReader(insurance_csv)
    young_dict = []
    for row in insurance_dict:
        if int(row['age']) <= 21:
            young_dict.append(row)
```

Let's see with what data we are working!

```python
for row in young_dict:
    print(row)
```

### Calculating median

After we have all young-aged people in a processable form, it is time to calculate the median of their insurance cost. Also, considering we are working with money values, it is better to work with Decimal variables than float ones.

```python
median_cost = 0
for row in young_dict:
    median_cost += Decimal(row['charges'])
median_cost /= len(young_dict)
print('The median cost of a young-aged person is {}.'.format(median_cost))
```

### Number of people with higher cost

Let's find out how many younger-aged people have their insurance cost higher than median!

```python
amount_higher = 0
for row in young_dict:
    if Decimal(row['charges']) > median_cost:
        amount_higher += 1
print('There are {} people in our dataset.'.format(len(young_dict)))
print('The amount of young people that have their insurance costs higher than average is {}.'.format(amount_higher))
print('Their percent of all young people is {}%.'.format(amount_higher / len(young_dict) * 100))
```
